```python
path = './results/tf-idf_subsample.json'
with open(path) as json_file:
    data = json.load(json_file)
# print(data)
MAP_scores = []
NDCG_scores = []
for key in data:
#     print(data[key])
    for score in data[key]:
        if score == 'map':
            MAP_scores.append(data[key][score])
        elif score == 'ndcg':
            NDCG_scores.append(data[key][score])
# print(MAP_scores)
# print(NDCG_scores)
mean_MAP = np.mean(MAP_scores)
mean_NDCG = np.mean(NDCG_scores)
print('MAP mean: ', mean_MAP)
print('NDCG mean: ', mean_NDCG)
```

```python
# ensure dataset is downloaded
# download_ap.download_dataset()
# pre-process the text
docs_by_id = read_ap.get_processed_docs()
print(len(list(docs_by_id.keys())))
# get qrels and queries
qrels, queries = read_ap.read_qrels()
```

```python
# divide docs into train and test corpus
train_docs = doc2vec.subsample_train(docs_by_id, train_docs_len=1000)
# save doc_ids and corresponding tags in dict
keys_tags_dict = {}
# convert docs into TaggedDocuments
train_corpus = list(doc2vec.read_corpus(train_docs, keys_tags_dict))
```

```python
print(len(train_docs))
```

```python
 # build model and train it
model, params = doc2vec.build_and_train_doc2vec(train_corpus)
# free up some ram
model.delete_temporary_training_data(keep_doctags_vectors=True, keep_inference=True)
```

```python
# try random query
random_qid = random.choice(list(queries.keys()))
random_query = queries[random_qid]
# get tf-idf scores for random query
results = doc2vec.doc2vec_search(random_query)
# print results of random query
print_results(docs_by_id, results, random_query)
```
